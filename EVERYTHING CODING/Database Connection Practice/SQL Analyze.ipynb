```python
query = '''CREATE TABLE IF NOT EXISTS chicago_socioeconomic_data 
(ca int PRIMARY KEY, 
community_area_name VARCHAR(40), 
percent_of_housing_crowded DECIMAL, 
percent_households_below_poverty DECIMAL, 
percent_aged_16_unemployed DECIMAL, 
percent_aged_25_without_high_school_diploma DECIMAL, 
percent_aged_under_18_or_over_64 DECIMAL, 
per_capita_income_ INT, 
hardship_index INT)'''
```

```python
db.execute('CREATE TABLE IF NOT EXISTS chicago_socioeconomic_data (cja int PRIMARY KEY, community_area_name VARCHAR(40), percent_of_housing_crowded DECIMAL, percent_households_below_poverty DECIMAL, percent_aged_16_unemployed DECIMAL, percent_aged_25_without_high_school_diploma DECIMAL, percent_aged_under_18_or_over_64 DECIMAL, per_capita_income_ INT, hardship_index INT)')
```

```python
chicago_socioeconomic_data.to_sql(con=engine,name='chicago_socioeconomic_data',if_exists='replace',index=False)
```

# Creating own table

```python
datas = {'product_name': ['Computer','Tablet','Monitor','Printer'],
        'price': [900,300,450,150]
        }
```

```python
products = pd.DataFrame(datas, columns= ['product_name','price'])
print (products)
```

```python
db.execute('CREATE TABLE IF NOT EXISTS products_df (product_name varchar(50), price int)')
```

```python
products.to_sql(con=engine,name='products_df',if_exists='replace',index=False)
```

```python
db.close()
```
