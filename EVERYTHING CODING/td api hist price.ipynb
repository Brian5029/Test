```python
import requests
import pandas as pd
import time
import json
from datetime import datetime
```

```python
end_date_string = '2022-01-01' #not including
start_date_string = '2019-01-01' #not including
# start_date_string < data < end_date_string

end_date_datetime = datetime.strptime(end_date_string, '%Y-%m-%d')
start_date_datetime = datetime.strptime(start_date_string, '%Y-%m-%d') 

end_date = int(end_date_datetime.timestamp()*1000)
start_date = int(start_date_datetime.timestamp()*1000)

print(end_date)
print(start_date)
```

```python
API_KEY = 'Q5G9Y3VALS4FZYSEHRVHMQAWQ0K4FOHW'
url = 'https://api.tdameritrade.com/v1/marketdata/TSLA/pricehistory'
response = requests.get(url, params = {'apikey': API_KEY,
                                       'periodType': 'ytd',
                                       'frequencyType': 'daily',
                                       'frequency': '1',
                                       'endDate': end_date,
                                       'startDate': start_date,
                                       'needExtendedHoursData': 'false'})
print(json.dumps(response.json(), indent = 1))
```

```python
raw_data = response.json()
raw_data
```

```python
pd.DataFrame(raw_data['candles'][0], index=[0]) #index = [0] is needed since values inside the dictionary is not in the list
```

```python
tsla_hist_price = pd.DataFrame()
for i in range(len(raw_data['candles'])):
    price_dict = raw_data['candles'][i]
    price_row = pd.DataFrame(price_dict, index = [i])
    price_row['datetime'] = pd.to_datetime(price_row['datetime'][i],unit='ms')
    tsla_hist_price = pd.concat([tsla_hist_price, price_row])
```
