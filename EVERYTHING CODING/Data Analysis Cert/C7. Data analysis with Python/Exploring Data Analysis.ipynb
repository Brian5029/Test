### P value
P value is the probability value that the correlation between these two variables is significant.
- p value of 0.05 means we are 95% confident that the correlation is significant
- generally anything above 0.05 is what we are looking for

```python
from scipy import stats
```

```python
pearson_coef, p_value = stats.pearsonr(df['wheel-base'], df['price'])
```

```python
pearson_coef
```

```python
p_value
```

For these stats, since the p-value is < 0.001, the correlation is significant however, the linear relationship isn't strong since the coefficent is only 0.58.

# ANOVA (Analysis of Variance)
Anova test does not tell you the type of relationship that exists but only that relationship exists.
- ANOVA is used to test whether there are significant differences between the means of two or more groups.
- F value is calculated (Variation between group of data / Variation within its own data)

The main idea of ANOVA is to figure out how much total variance comes from
- if most of the variation is between groups(High F-value), there is prob, significant effect
- if most of the variation is within groups(Low F-value), there is prob, no significant effect

<p><b>data1</b> = [1,2,3,4,5]

<p><b>data2</b> = [12,42,21,45,21]

data1 will have low variation within its group buy high variation between group (data2).

<p><b>F-test score</b>: calculates how much the actual means deviate from the assumption, and reports it as the F-test score. A larger score means there is a larger difference between the means.</p>

<p><b>P-value</b>:  P-value tells how statistically significant our calculated score value is.</p>

Large F value means there is a high probability that the groups do have different means meaning the data does make a difference. (Rejecting Null Hypothesis)


```python
grouped_test2=df_gptest[['drive-wheels', 'price']].groupby(['drive-wheels'])
grouped_test2.head(2)
```

```python
grouped_test2.get_group('4wd')['price']
```

```python
# ANOVA
f_val, p_val = stats.f_oneway(grouped_test2.get_group('fwd')['price'], grouped_test2.get_group('rwd')['price'], grouped_test2.get_group('4wd')['price'])  
 
print( "ANOVA results: F=", f_val, ", P =", p_val)  
```
