# **Types of SQL Statement**
- DDL (Data Definition Language) statement define, change, or drop data
  - CREATE: creating tables and definding column
  - ALTER: altering tables including adding, dropping and, modifying datas
  - TRUNCATE: deleting data in the table but not the table itself
  - DROP: deleting tables
- DML (Data Manipulation Language) statement read and modify data
  - INSERT: used for inserting row into the table
  - SELECT: reads or selects row from the table
  - UPDATE: edit rows from the table
  - DELETE: removes a row of data from a table

# **ALTER, DROP, and TRUNCATE tables**
- How does the syntax of a CREATE statement look?


    CREATE TABLE table_name (

    column1 datatype,

    column2 datatype,

    column3 datatype,

    ....

    );

- How does the syntax of an ALTER statement look?


    ALTER TABLE table_name
    ADD COLUMN column_name data_type column_constraint;

    ALTER TABLE table_name
    DROP COLUMN column_name;

    ALTER TABLE table_name
    ALTER COLUMN column_name SET DATA TYPE data_type;

    ALTER TABLE table_name
    RENAME COLUMN current_column_name TO new_column_name;

- How does the syntax of a TRUNCATE statement look?


    TRUNCATE TABLE table_name;

- How does the syntax of a DROP statement look?


    DROP TABLE table_name;
