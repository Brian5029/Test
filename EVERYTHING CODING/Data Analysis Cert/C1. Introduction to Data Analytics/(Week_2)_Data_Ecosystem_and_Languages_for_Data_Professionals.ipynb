# **Types of Data**

- Structured Data has a defined structure and can be represented with rows and columns
  - Numbers that can be collected, exported, stored.
- Semi-Structured Data has some organizational properties but not all. It cannot be stored in the forms of rows and columns
  - E-mail, Zipped files, etc
  - XML/JSON are used to store these datas
-Unstructured Data does not have easily identifiable structure and does not follow any particular format.
  - images, video, documents
  - can be stored in files such as word document



# **Different Types of File Formats**

- Delimited text files values are separated by a delimiter such as comma (CSV) or tab (TSV).
- XLSX is a spreadsheet format created by Microsoft. Basically datas stored in Microsoft Excel
- XML is a markup language with set rules that can be readable for humans and machines.
- PDF is a portable document independent of application software and hardware. It can be viewed anywhere with human eyes.
- JSON is a text based designed for transfering data over the web. It can be read in any programming language 

# **Sources of Data**

- Relational Databases has internal database for Business, Transactions, Workflows
- XML Datasets
  - Government Documents, Private company's financial, point of sale.
- Flat files, data stored in plain text and each line or row is one data record. Separated by a delimiter (CSV, XLSX)
- XML can support complex data structure such as online surveys, bank statement, etc.
- API is used to extract data from website

# **Data Repository**
Data Repository is a term used to refer data that has been collected and organized.

Types of data repositories
- Data bases is a collection of data for input, storage, search, retrievel of data.
  - Database is designed to record data
  - Database uses Online Transactional Processing (OLTP)
- Data Warehouse gathers all datas from different source into comprehensive database.
  - Data Warehouse is designed to analyze data
  - Data warehouse uses Online Analytical Processing (OLAP)
- Datas that are in Data Base goes through `ETL (Extract, Transform, Load)` Process to Data Warehouse where datas can be analyized




# **RDBMS (Relational Database Management Systems)** 
relational database is a collection of data that are related and organized into table strucutre.

- Relational data base is optimal for large storage, processing of data while spreadsheet's space is limited.
- RDBMS databases can be open-source and closed and can be supported by cloud service like Amazon RDS or MySQL

Advantages of Relational Databases
- Create meaningful info by joining tables
- Flexibility
- Offer export and import options for easy backup or recovery of data

Disadvatages of Relational Databases
- Does not work with semi-structured and unstructured data
- Migration of RDBMS data is ONLY possible when the destination tables are identical
- Entering value greater than defined length will delete excess datas

# **NoSQL**
NoSQL is a non-relational database designed to provide flexible schemas for storage.

- Does not use SQL to query data
- Does not use traditional row/column/table database with fixed schemas
