# Remove Empty Rows
- `CTR+HOME` to select everything
- `CTR+END` to go back to start
- Clicking Filter in data tab after selecting columns will give filter dropdown options 

# Deleting Duplicates
- Select the column and click (under home) Conditional Formatting> Highlight Cells Rules> Duplicate Values
- Make sure the column you are selecting is expected to not have any duplicates. For example car model names
- This will highlight all duplicates and can delete manually

OR
- Select all datas and click Remove duplicates under Data tab
- This is used when trying to find duplicate rows. Same values for every column. It compares all columns rather than 1

# Find & Replace
- Find & Select under Home tab

# Dealing with Inconsistencies in data (Proper,Upper,Lower case)
- Insert new row or column under where the data case is going to change
- =PROPER,UPPER,LOWER(A#) will change the data to Proper case
- cick the first cell, select all rows `ctr+shift+right arrow`
- press f2 (for rows)
- press ctr enter (for rows)
- After copy and choose paste options Values.

# Change date format
- Select column and click Number Format > More Number Formats and Date under Home

# Trim White Space
- Under Find & Select, type 2 space for find and type 1 space for replace
ALSO
- Can use =TRIM(A#) function to trim white space. Same process as dealing with inconsistencies.

# Flash Fill
Flash fill will autofill the data that uses two column values in pattern. For example
- Column A = Custumer name 
- Column B = First name
- Column C = Last name
- Typing in data column B and C value in column A and clicking flash fill will fill every rows of the column corresponding to column B and C

# Text to Column (Splitting words into columns)
- Insert new columns for new values to be put in
- Select the column that will be splitting
- Click text to column in data
- delimited, select delimiters, and destination of the data.
