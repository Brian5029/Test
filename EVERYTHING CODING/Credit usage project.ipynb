## Removing Paycheck from Category column and createing its own dataframe

Lets find the indicies where `Category` is equal to Paycheck

```python
paycheck_condition = seasons_added_df['Category']=='Paycheck'
paycheck_condition
```

```python
raw_index = seasons_added_df.index
raw_index
```

```python
paycheck_indicies = raw_index[paycheck_condition]
paycheck_indicies
```

```python
paycheck = paycheck_indicies.tolist()
paycheck
```

### Creating Paycheck only dataframe

```python
paycheck_df = pd.DataFrame()
for paycheck_index in paycheck:
    paycheck_df =  paycheck_df.append(seasons_added_df.loc[paycheck_index])

#Resetting index
paycheck_df.reset_index(inplace = True)
paycheck_df.drop(columns= ['index'], inplace = True)
paycheck_df
```

### Removing `Paycheck` from `Category` column

```python
spending_transaction_df = seasons_added_df.copy()
for deposit in paycheck:
    spending_transaction_df = spending_transaction_df.drop([deposit])
spending_transaction_df = spending_transaction_df.reset_index(drop=True)
spending_transaction_df
```

To confirm if deposit column was removed we compare the length of the df

```python
spending_transaction_df.loc[spending_transaction_df['Category']=='Paycheck']
```
